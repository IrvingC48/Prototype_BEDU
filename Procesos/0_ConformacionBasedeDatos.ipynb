# Estandarización de Valores Horarios

## Conversión de horas

Una vez los datasets fueron visualizados, se observó que los datos horarios de los contaminantes iban desde la hora 1 a la hora 24; siendo diferente este formato con respecto a los datos meteorológicos.

Por esta razón, fue necesario transformar el formato del dataset de los contaminantes para lo que se siguió el siguiente procedimiento.

```python
import datetime as dt

def my_to_datetime(date_str):
    if date_str[11:13] != '24': #Identificar las que no son la hora 24 y guardar como tal
        return pd.to_datetime(date_str, format='%d/%m/%Y %H:%M')

    date_str = date_str[0:11] + '00' + date_str[13:] #Tranformar los que son la hora 24 y devolver la transformación para la hora 0 del siguiente día
    return pd.to_datetime(date_str, format='%d/%m/%Y %H:%M') + \
           dt.timedelta(days=1)
```

```python
datos_completos_pm['date']=datos_completos_pm.date.apply(my_to_datetime) #Aplicar la función a la columna
```

```python
datos_completos_pm.tail()
```
